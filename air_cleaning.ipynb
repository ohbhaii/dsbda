```python
import pandas as pd
```

```python
data = pd.read_csv('/kaggle/input/air-quality-dataset/data.csv',sep=',')
```

```python
data.head()
```

```python
#Data Cleaning
data.isnull()
```

```python
data.dropna(subset=['date'])
```

```python
data.columns
```

```python
data.drop(['so2', 'no2', 'rspm', 'spm'],axis=1)
```

```python
data.fillna(0)
```

```python
#Data Integration
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
data.columns
```

```python
columns = ['so2', 'no2', 'rspm', 'spm']
```

```python
imp = SimpleImputer(missing_values=np.NaN, strategy='mean')
```

```python
data[columns] = imp.fit_transform(data[columns])
```

```python
data[columns]
```

```python
data[columns].isnull()
```

```python
#Data Transformation
data.columns
```

```python
data['type']
```

```python
data.head()
```

```python
#categorical to categorical
data['type'].replace({'Residential, Rural and other Areas':'RRA','Industrial Area':'IA','RIRUO':'RIR'},inplace=True)
```

```python
#categorical to numerical
data['type'].replace({'RRA':1,'IA':2,'RIR':3})
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
data['state'] = label.fit_transform(data['state'])
```
