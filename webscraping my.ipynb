```python
request1.content
```

```python
soup=bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews=soup.findAll('div', {'class' : 't-ZTKy'})
for review in reviews:
    print(review.get_text() + "\n\n")
```

```python
rating=soup.find('div', {'class' : '_2d4LTz'}).get_text()
print(rating)
```

```python
ratings=soup.findAll('div',{'class' : '_3LWZlK _1BLPMq'})
for rating in ratings:
    print(rating.get_text())
```

```python
questions=soup.findAll('div', {'class': '_1xR0kG _3cziW5'})
for question in questions:
    print(question.get_text())
answers=soup.findAll('div', {'class': '_2yeNfb'})
for answer in answers:
    print(answer.get_text())
    
```
