```python
# import pandas library
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Reading csv file
data = pd.read_csv("dataset_Facebook.csv")
# creating a dataframe of it
data = pd.DataFrame(data)
data.head()
```

```python
# Datatype of each column
data.dtypes
```

```python
# Information about each column data
data.info()
```

```python
# Count ,min,max ,etc of each column
data.describe()
```

```python
#Finding null values in each column
data.isna().sum()
```

```python
# Replacing null values of Paid with it's mean
paid_mean = data['Paid'].mean()
print(paid_mean)
data['Paid'].fillna(paid_mean, inplace=True)
data.isna().sum()
```

```python
# Replacing null values of like with it's mean
like_mean = data['like'].mean()
print(like_mean)
data['like'].fillna(like_mean, inplace=True)
data.isna().sum()
```

```python
# Replacing null values of Paid with it's mean
share_mean = data['share'].mean()
print(share_mean)
data['share'].fillna(share_mean, inplace=True)
data.isna().sum()
```

```python
#Finding number of zeros in each column
for i in data:
  count = (data[i] == 0).sum()
  print('Zeros in column',i,'->', count)
```

```python
# splitting data using columns
data_frame1=data[['Page total likes','Type','Category','Post Month','Post Weekday','Post Hour', 'comment', 'like', 'share']]
data_frame1.head()
```

```python
#splitting data using loc function
data_frame2=data.loc[:,['Page total likes','Paid']]
data_frame2.head()
```

```python
# splitting data using iloc function using indexes of columns
data_frame3=data.iloc[:,[0,15,16,17]]
data_frame3.head()
```

```python
# splitting data using train test split
x = data.drop(['Paid'], axis=1)
y = data.Paid
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

x_train.head(),x_test.head(),y_train.head(),y_test.head()
```

```python
# Merging data_frame1 and data_frame2
data_frame3=pd.merge(data_frame1, data_frame2)
data_frame3.head()
```
